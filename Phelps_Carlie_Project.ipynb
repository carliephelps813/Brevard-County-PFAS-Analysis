```python
## DATA DESCRIPTION ##


import sys
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA


## Atlantic Coast ##
file = open(input()) # Open Atlantic Coast PFAS.txt
conc19_list_atl = []
conc21_list_atl = []
types_19_atl = 0
types_21_atl = 0

for line in file:
    line = line.strip()
    data = line.split("\t")
    if data[1] == "2019":
        conc19_list_atl.append(float(data[13]))
        if data[13] != "0":
            types_19_atl = types_19_atl + 1
    elif data[0] == "2021":
        conc21_list_atl.append(float(data[12]))
        if data[12] != "0":
            types_21_atl = types_21_atl + 1
# Means
mean_19_atl = round(np.mean(conc19_list_atl), 2)
mean_21_atl = round(np.mean(conc21_list_atl), 2)

# Medians
med_19_atl = round(np.median(conc19_list_atl), 2)
med_21_atl = round(np.median(conc21_list_atl), 2)

# Standard Deviations
st_19_atl = round(np.std(conc19_list_atl), 2)
st_21_atl = round(np.std(conc21_list_atl), 2)

print("Mean, Median, Standard Deviation for Atlantic Coast 2019:", mean_19_atl, med_19_atl, st_19_atl)
print("Mean, Median, Standard Deviation for Atlantic Coast 2021:", mean_21_atl, med_21_atl, st_21_atl)


## Banana River ##
file = open(input()) # Open Banana River PFAS.txt
conc19_list_br = []
conc21_list_br = []
types_19_br = 0
types_21_br = 0

for line in file:
    line = line.strip()
    data = line.split("\t")
    if data[1] == "2019":
        conc19_list_br.append(float(data[15]))
        if data[15] != "0":
            types_19_br = types_19_br + 1
    elif data[0] == "2021":
        conc21_list_br.append(float(data[14]))
        if data[14] != "0":
            types_21_br = types_21_br + 1
# Means
mean_19_br = round(np.mean(conc19_list_br), 2)
mean_21_br = round(np.mean(conc21_list_br), 2)

# Medians
med_19_br = round(np.median(conc19_list_br), 2)
med_21_br = round(np.median(conc21_list_br), 2)

# Standard Deviations
st_19_br = round(np.std(conc19_list_br), 2)
st_21_br = round(np.std(conc21_list_br), 2)

print("Mean, Median, Standard Deviation for Banana River 2019:", mean_19_br, med_19_br, st_19_br)
print("Mean, Median, Standard Deviation for Banana River 2021:", mean_21_br, med_21_br, st_21_br)


## North Indian River ##
file = open(input()) # Open N Indian River PFAS.txt
conc19_list_nir = []
conc21_list_nir = []
types_19_nir = 0
types_21_nir = 0

for line in file:
    line = line.strip()
    data = line.split("\t")
    if data[1] == "2019":
        conc19_list_nir.append(float(data[11]))
        if data[11] != "0":
            types_19_nir = types_19_nir + 1
    elif data[0] == "2021":
        conc21_list_nir.append(float(data[10]))
        if data[10] != "0":
            types_21_nir = types_21_nir + 1
# Means
mean_19_nir = round(np.mean(conc19_list_nir), 2)
mean_21_nir = round(np.mean(conc21_list_nir), 2)

# Medians
med_19_nir = round(np.median(conc19_list_nir), 2)
med_21_nir = round(np.median(conc21_list_nir), 2)

# Standard Deviations
st_19_nir = round(np.std(conc19_list_nir), 2)
st_21_nir = round(np.std(conc21_list_nir), 2)

print("Mean, Median, Standard Deviation for N Indian River 2019:", mean_19_nir, med_19_nir, st_19_nir)
print("Mean, Median, Standard Deviation for N Indian River 2021:", mean_21_nir, med_21_nir, st_21_nir)


## South Indian River ##
file = open(input()) # Open S Indian River PFAS.txt
conc19_list_sir = []
conc21_list_sir = []
types_19_sir = 0
types_21_sir = 0

for line in file:
    line = line.strip()
    data = line.split("\t")
    if data[1] == "2019":
        conc19_list_sir.append(float(data[9]))
        if data[9] != "0":
            types_19_sir = types_19_sir + 1
    elif data[0] == "2021":
        conc21_list_sir.append(float(data[8]))
        if data[8] != "0":
            types_21_sir = types_21_sir + 1
# Means
mean_19_sir = round(np.mean(conc19_list_sir), 2)
mean_21_sir = round(np.mean(conc21_list_sir), 2)

# Medians
med_19_sir = round(np.median(conc19_list_sir), 2)
med_21_sir = round(np.median(conc21_list_sir), 2)

# Standard Deviations
st_19_sir = round(np.std(conc19_list_sir), 2)
st_21_sir = round(np.std(conc21_list_sir), 2)

print("Mean, Median, Standard Deviation for S Indian River 2019:", mean_19_sir, med_19_sir, st_19_sir)
print("Mean, Median, Standard Deviation for S Indian River 2021:", mean_21_sir, med_21_sir, st_21_sir)


# Histogram: Distribution of Mean PFAS Concentrations Overall
means_list = [mean_19_atl, mean_21_atl, mean_19_br, mean_21_br, mean_19_nir, mean_21_nir, mean_19_sir, mean_21_sir]
plt.hist(means_list, bins = 4, edgecolor = 'black')
title = "Distribution of Mean PFAS Concentrations in Brevard County Sites, 2019 and 2021"
plt.title(title)
plt.xlabel('PFAS Concentration (ng/L)')
plt.ylabel('Number of Mean Concentrations in Range')

# Histogram: Distriubtion of PFAS Variation Overall
plt.figure()
types_list = [types_19_atl, types_21_atl, types_19_br, types_21_br, types_19_nir, types_21_nir, types_19_sir, types_21_sir]
plt.hist(types_list, bins = 'auto', edgecolor = 'black')
title = "Distribution of PFAS Variation in Brevard County Sites, 2019 and 2021"
plt.title(title)
plt.xlabel('Number of PFAS Types Present')
plt.ylabel('Number of Sites in Range')
```

```python
## CORRELATION ##


# Pearson and Spearman Coefficients for Concentrations
rel_dev_list = [50, 50, 80, 80, 60, 60, 80, 80] # Estimated percent of development around the edges of each location
mean_pearson = stats.pearsonr(means_list, rel_dev_list)
mean_spear = stats.spearmanr(means_list, rel_dev_list)
print("PFAS Concentrations vs Development, Pearson =", round(mean_pearson.statistic, 3), "Spearman =", round(mean_spear.statistic, 3))

# Pearson and Spearman Coefficients for Types
types_pearson = stats.pearsonr(types_list, rel_dev_list)
types_spear = stats.spearmanr(types_list, rel_dev_list)
print("PFAS Types vs Development, Pearson =", round(types_pearson.statistic, 3), "Spearman =", round(types_spear.statistic, 3))

# Linear Regression for Concentrations
plt.figure()
slope, intercept = np.polyfit(rel_dev_list, means_list, 1)
Y_pred = np.polyval([slope, intercept], rel_dev_list)
plt.scatter(rel_dev_list, means_list, color = 'blue', label = 'Actual data')
plt.plot(rel_dev_list, Y_pred, color = 'red', label = 'Regression line')
plt.xlim(0, 100)
plt.ylim(0, 100)
title = "Linear Regression for Mean PFAS Concentrations in Brevard County Sites, 2019 and 2021"
plt.title(title)
plt.xlabel('Percent Development on Location Edges')
plt.ylabel('PFAS Concentration (ng/L)')
plt.legend()

# Linear Regression for Types
plt.figure()
slope, intercept = np.polyfit(rel_dev_list, types_list, 1)
Y_pred = np.polyval([slope, intercept], rel_dev_list)
plt.scatter(rel_dev_list, types_list, color = 'blue', label = 'Actual data')
plt.plot(rel_dev_list, Y_pred, color = 'red', label = 'Regression line')
plt.xlim(0, 100)
plt.ylim(10, 20)
title = "Linear Regression for PFAS Types in Brevard County Sites, 2019 and 2021"
plt.title(title)
plt.xlabel('Percent Development on Location Edges')
plt.ylabel('Number of PFAS Types Present')
plt.legend()
```
